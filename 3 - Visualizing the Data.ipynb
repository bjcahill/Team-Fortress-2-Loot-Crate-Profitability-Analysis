```python
%matplotlib inline
import numpy as np
import pandas as pd
from altair import *
pd.options.display.max_rows = 10
```

```python
unusuals_df = pd.read_csv("CompleteHatAndPricingData.csv")
unusuals_df = unusuals_df[unusuals_df.columns.drop("Unnamed: 0")]
unusuals_df
```

## Visualizing the Types of Unusuals

```python
from altair import *

class_domain = ["Scout","Soldier","Pyro","Demoman","Heavy","Engineer","Medic","Sniper","All classes","Multi-Class"]
class_range_ = ['#194FAE', '#194FAE', '#194FAE','#194FAE','#194FAE','#194FAE','#194FAE','#194FAE','#EE8130','#FFCE00']

data_transformers.enable('json')
Chart(unusuals_df).mark_circle(size = 60).encode(
    x=X("Effect Community Rating"),
    y=Y("Price", scale=Scale(type='log',base = 2)),
    color=Color('Class', scale=Scale(domain = class_domain,range = class_range_)),
    column="Item Type"
).configure_mark(
    opacity=0.6)
```

## Visualizing the Types of Unusual Hats

```python
cosmetics_df = unusuals_df[unusuals_df["Item Type"] == "Cosmetic"]
```

```python
%%capture --no-stdout
cosmetics_df ["Type of Cosmetic Unusual"] = cosmetics_df.apply(
    lambda row: "Category 1: Robo" if row["Robo"] == True else "Category 2: Regular",
    axis=1)
cosmetics_df ["Type of Cosmetic Unusual"] = cosmetics_df.apply(
    lambda row: "Category 3: Misc" if row["Equip Region"] == "Misc" and row["Type of Cosmetic Unusual"] == "Category 2: Regular" else row["Type of Cosmetic Unusual"],
    axis=1)
```

```python
domain = ["Mercenary","No Grade","Commando","Assassin","Elite"]
range_ = ['#194FAE','#56AC4D','#582685', '#FF00A7','#FF0000']

data_transformers.enable('json')
Chart(cosmetics_df).mark_circle().encode(
    x=X("Effect Community Rating"),
    y=Y("Price", scale=Scale(type='log',base = 2)),
    color=Color('Grade', scale=Scale(domain = domain,range = range_)),
    column="Type of Cosmetic Unusual",
    size = Size("Grade",scale=Scale(domain = ["Mercenary","No Grade","Commando","Assassin","Elite"], range = [50,200,400,8000,10000]))
).configure_mark(
    opacity=0.6)
```

## Hats vs Type of Crate

```python
unusuals_df["Type of Crate Temp"] = unusuals_df["Unboxed From"].str.contains("Crate")
unusuals_df["Type of Crate"] = unusuals_df.apply(
    lambda row: "Category 1: Unlocked Crate" if row["Type of Crate Temp"] == True else "Category 2: Other Series",
    axis=1)

unusuals_df["Type of Crate"].value_counts()
```

```python
median_prices = unusuals_df.groupby("Name")["Price"].median().to_frame().reset_index().rename(columns = {"Price": "Median Hat Price"})
unique_hats_df = unusuals_df.merge(median_prices,on = "Name",how = "left").drop_duplicates("Name")
```

```python
Chart(unique_hats_df[unique_hats_df["Median Hat Rating"] < 25]).mark_circle(size = 60).encode(
    x=X("Median Hat Rating"),
    y=Y("Median Hat Price", scale=Scale(type='log',base = 2)),
    color=Color('Grade', scale=Scale(domain = domain,range = range_)),
    column="Type of Crate",
    size = Size("Grade",scale=Scale(domain = ["Mercenary","No Grade","Commando","Assassin","Elite"], range = [50,200,400,8000,10000]))
).configure_mark(
    opacity=0.5)
```

## Effect Generation vs Price 

```python
median_prices = unusuals_df.groupby("Name")["Price"].median().to_frame().reset_index().rename(columns = {"Price": "Median Hat Price"})
unique_effects_df = unusuals_df.merge(median_prices,on = "Name",how = "left")
unique_effects_df = unique_effects_df[~unique_effects_df["Price"].isnull()]
unique_effects_df = unique_effects_df.drop_duplicates("Effect")
```

```python
halloween_generations = ["Halloween 2011","Halloween 2012","Halloween 2013","Halloween 2014","Halloween 2015","Halloween 2016","Halloween 2018"]
unique_effects_df["Halloween Effect"] = unique_effects_df.apply(
    lambda row: "Halloween" if row["Generation"] in halloween_generations else "Regular",
    axis=1)
```

```python
Chart(unique_effects_df).mark_circle(size = 80).encode(
    x=X("Effect Community Rating"),
    y=Y("Effect Median Price"),
    color="Halloween Effect",
).configure_mark(
    opacity=0.8)
```
